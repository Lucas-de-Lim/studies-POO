## Interfaces

Interfaces são métodos ou atributos públicos que outros objetos possam usar para se comunicar com outros objetos 

Explanação sobre lista e container.. A lista se comporta como um container, mas não é um container. Sera que ela é uma subclasse de container?

```python
class Time:
    def __init__(self, membros):
        self._membros = membros
    
    def __len__(self):
        return len(self._membros)
    
    def __contains__(self, member):
        return member in self.__membros
```

```python
Membros = ['Lucas', 'Felipe', 'Jorge']

time = Time(Membros)
```

```python
len(time)
```

```python
class Trem:
    def __init__(self, vagoes):
        self._vagoes = vagoes
    
    def __len__(self):
        return len(self._vagoes)
    
    def __contains__(self, vagao):
        return vagao in self._vagoes
```

```python
vagoes = range(0,358)
```

```python
trem = Trem(vagoes)
```

```python
288 in trem
```

Não pode instanciar essa classe que herda de collection pois não implementa esses dunders que caracterizam a Collection

```python
class Container(Collection):
    
    def __init__(self, seq):
        self.seq = seq
    
    def __contains__(self, outro):
        return outro in list(self.seq)
        
    def __len__(self):
        return len(self.seq)
    
    def __iter__(self):
        return  self
    
```

```python
c = Container(time)
```
