## Composição

é a maneira de ligar objetos que não tem herança entre si

```python
class Pizzaroa():
    
    def __init__(self):
        self._forno = Forno()
        
        
    def pedido(self, pizza):
        self._forno.assar(pizza)

    
```

```python
class Forno():
    
    def __init__(self):
        self.pizzas = []
        self.lenha = None
        
    def assar(self, pizza):
        if not self.lenha:
            print('Não há lenha')
```

Deixando as coisas mais cabulosas, tudo para entender a resposabilidade dos atores e as inteações entre si.

```python
class Pizzaria():
    
    def __init__(self):
        self._pizzaolo = Pizzaolo
        
        
        
    def pedido(self, pizza):
        self._pizzaolo.preparar_pizza(pizza)
        
        
class Forno():
    
    def __init__(self):
        self.pizzas = []
        self.lenha = None
        
    def assar(self, pizza):
        if not self.lenha:
            print('Não há lenha')
            
            
class Pizzaolo():
    
    def __init__(self):
        self._forno = Forno()
        
    
    def preparar_pizza(self, pizza):
        self._forno.assar(pizza)
```

```python
pizzaria = Pizzaria()
```

```python
pizzaria.pedido('mussarela')
```

## Primeiros Dunders

metodos especiais

anotação: toda função/metodo que nao retorna nada, na verdade retorne None

```python
class MinhaClasse():
    
    def __new__(cls):
        print('construindo')
        
        return super().__new__(cls)
        
    def __init__(self):
        print('inicializando')
        
        
    def __del__(self):
        print('finalizando')
```

```python
a = MinhaClasse()
```

```python
class Passaro:
    
    def __init__(self, nome):
        self.nome = nome
        
        
class CanarinhoPistola(Passaro):
    
    def __init__(self,nome, camisa):
        super().__init__(none)
        self.camisa = camisa
        
    def __str__(self):
        return 'CanarinhoPistola({nome="self.nome"}, camisa="{camisa.self}")'
        
```
