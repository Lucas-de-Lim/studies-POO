## Sobrecarga de operadores

```python
class Dois:
    val = 2
```

```python
Dois.val
```

```python
d = Dois()
```

```python
-d
```

Operador unário pelo que entendi são esses normais da linguagem + - ~ que ficam na frente do numero

Mas aí quando você quer misturar uma classe com ele, bagulho da zika aí tem que fazer essa sobrecarga de operador

```python
class Dois:
    val = 2
    
    def __neg__(self):
        print('negativei')
        return -self.val
    
    def __pos__(self):
        print('positivei')
        return +self.val
```

```python
d = Dois()
```

```python
-d
```

```python
+d
```
